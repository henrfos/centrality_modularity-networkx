```python
import networkx as nx
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('nutrients.csv', header=None)
G = nx.from_pandas_edgelist(df, 0, 1)
```

```python
deg_cen = nx.degree_centrality(G)
eig_cen = nx.eigenvector_centrality(G)
```

```python
modularity = nx.algorithms.community.modularity(G, nx.algorithms.community.label_propagation_communities(G))
```

```python
node_sizes = [deg_cen[node] * 5000 for node in G.nodes()]
node_colors = [eig_cen[node] for node in G.nodes()]
```

```python
plt.figure(figsize=(12, 12))
pos = nx.spring_layout(G)
nx.draw_networkx_nodes(G, pos, node_size=node_sizes, node_color=node_colors, cmap=plt.cm.Blues)
nx.draw_networkx_edges(G, pos)
nx.draw_networkx_labels(G, pos, font_size=10, font_family='sans-serif')
plt.axis('off')
plt.show()
```
